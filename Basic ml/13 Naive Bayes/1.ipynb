```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head() #these parameters are not necessary.
```

```python
target = df.Survived
inputs= df.drop('Survived',axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex,dtype=int)
dummies.head()
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs= inputs.drop('Sex',axis='columns')
inputs.head()
```

```python
inputs.columns[inputs.isnull().any()]

#This line finds if there are any NaN or null values in the inputs dataframe
```

```python
inputs.Age[:10]
```

```python
x = inputs.Age.mean()
inputs.Age=inputs.Age.fillna(x)

#fills the NaN with the average Age value
```

```python
inputs.head(7)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```
