```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
dft=df
dft['target']=digits.target
dft.head()
```

```python
df.drop(['target'],axis='columns')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(df,digits.target,test_size=0.2)
```

```python
len(X_test)
```

ensemble  is a thing when you use multiple models to predict. here we are using a lot of decision trees to predict which combined is called random forest

 sklearn requires all feature names (column names of the input X_train) to be strings. However, your dataset contains mixed types, such as integers or other data types, in the feature names.

To solve this issue, ensure that all column names in your input data (X_train) are strings. Here's how you can fix it:

```python

X_train.columns = X_train.columns.map(str)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train,Y_train)
```

```python
X_test.columns = X_test.columns.map(str)
```

```python
model.score(X_test,Y_test)
```

```python
Y_predicted= model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predicted)
cm
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 7))

# Plot a heatmap using seaborn
sns.heatmap(cm, annot=True)

# Set labels for the x-axis and y-axis
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
